# Prepare EGEDA data for Agriculture model

```python
import pandas as pd
import numpy as np
```

```python
df_EGEDA = pd.read_csv("../data/final/EGEDA_2019_09_25_tidy.csv",
                      index_col=['Economy'])
```

```python
df_EGEDA.head()
```

```python
df_EGEDA.columns.values.tolist()
```

```python
df_EGEDA['Fuel Code'].unique()
```

```python
top_fuels = ['Coal','Elec','Gas','Heat','Nuc','Oil','Oth','PetP','RenB','RenG','RenH','RenMSW','RenNRE','RenO','RenS','RenW','Tot']
```

```python
ag_df = df_EGEDA[['Year','Fuel Code','15.2 Agriculture']]
ag_df.head()
```

```python
ag_df.tail()
```

```python
ag_df.info()
```

```python
ag_table = pd.pivot_table(ag_df, values='15.2 Agriculture', index=['Economy', 'Year'], columns=['Fuel Code'], aggfunc=np.sum)
```

```python
ag_table.loc['AUS']
```

```python
ag_table[top_fuels].loc['AUS']
```

```python
ag_table[top_fuels].to_csv("../data/intermediate/ag_df.csv")
```
